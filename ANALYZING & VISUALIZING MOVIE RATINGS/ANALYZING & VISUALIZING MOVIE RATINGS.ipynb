```python
data = {
    "Movie": ["Movie A", "Movie B", "Movie C", "Movie D", "Movie E", "Movie F"],
    "Genre": ["Action", "Drama", "Comedy", "Action", "Drama", "Comedy"],
    "Rating": [8.5, 7.2, 6.8, 9.1, 8.0, 7.5]
}

df = pd.DataFrame(data)
```

```python
print(df)
```

```python
print("\nSummary Statistics:")
print(df["Rating"].describe())
```

```python
print("\nMean Rating:", df["Rating"].mean())
print("Median Rating:", df["Rating"].median())
print("Mode Rating:", df["Rating"].mode()[0])
```

```python
plt.hist(df["Rating"])
plt.title("Rating Distribution")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

```python
sns.boxplot(x=df["Rating"])
plt.title("Boxplot of Movie Ratings")
plt.show()
```

```python
top_movie = df[df["Rating"] == df["Rating"].max()]
print("\nTop Rated Movie:\n", top_movie)
```

```python
genre_rating = df.groupby("Genre")["Rating"].mean()
print("\nGenre-wise Average Rating:\n", genre_rating)

genre_rating.plot(kind="bar")
plt.title("Genre-wise Average Rating")
plt.xlabel("Genre")
plt.ylabel("Average Rating")
plt.show()
```
